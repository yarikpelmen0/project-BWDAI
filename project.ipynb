# **Загрузка и первичный обзор**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("anime-dataset-2023.csv")

display('Первые и последние пять строк:')
display(df.head())
display(df.tail())
```

```python
display('Базовая информация:')
display(df.info())
display(f'Кол-во строк и столбцов: {df.shape}')
```

# Обработка данных

```python
display('Количество пропусков:')
df.isnull().sum()
```

```python
display(f'Количество дубликатов: {df.duplicated().sum()}')
```

```python
df['Episodes'] = pd.to_numeric(df['Episodes'], errors='coerce')
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
df['Rank'] = pd.to_numeric(df['Rank'], errors='coerce')
```

# Статистики данных

```python
df.describe()
```

```python
print('Кол-во уникальных названий:')
print(df['Name'].nunique())
print(df['Name'].unique())

print('Кол-во уникальных комбинаций жанров:')
print(df['Genres'].nunique())
print(df['Genres'].unique())
```

```python
corr = df.corr(numeric_only=True)
corr
```

# Визуализация данных

```python
df['Rating'].hist(bins=20, xrot=45, xlabelsize=7)
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Score'].dropna(), fill=True)
plt.title('Распределение оценок аниме')
plt.xlabel('Оценка')
plt.ylabel('Плотность')
plt.grid(True)
plt.show()
```

```python
df_filtered = df.dropna(subset=['Score', 'Popularity'])

plt.figure(figsize=(15, 7))
sns.lineplot(x='Popularity', y='Score', data=df_filtered)
plt.title('Оценка аниме в зависимости от популярности')
plt.xlabel('Популярность')
plt.ylabel('Оценка')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
df['Type'].value_counts().plot(kind='bar')
plt.title('Распределение аниме по типу')
plt.xlabel('Тип Аниме')
plt.ylabel('Количество аниме')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['Score'].dropna())
plt.title('Распределение оценок аниме')
plt.ylabel('Оценка')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Группировки и агрегации

```python
grouped_data = df.groupby('Type').agg(
    mean_score=('Score', 'mean'),
    anime_count=('anime_id', 'count'),
    total_members=('Members', 'sum')
).reset_index()

display('Средняя оценка по типу аниме:')
display(grouped_data[['Type', 'mean_score']].sort_values(by='mean_score', ascending=False))

display('Количество аниме по типу:')
display(grouped_data[['Type', 'anime_count']].sort_values(by='anime_count', ascending=False))

display('Общее количество участников по типу аниме:')
display(grouped_data[['Type', 'total_members']].sort_values(by='total_members', ascending=False))
```

```python
pivot_table = df.pivot_table(index='Type', columns='Rating', values='Score', aggfunc='mean')

display('Сводная таблица средних оценок:')
display(pivot_table)
```
